1. Задание (в программе)
Нарисуйте график функции:
y(x) = k∙cos(x – a) + b
для некоторых (2-3 различных) значений параметров k, a, b


```python
import math
import matplotlib.pyplot as plt
import decimal
import numpy as np
from numpy import (linspace,logspace,zeros,ones,outer,meshgrid,
                   pi,sin,cos,sqrt,exp)
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
from math import pi
import random

varmax = 20
varmin = -20

k=random.randint(varmin,varmax)
k1=random.randint(varmin,varmax)
k2=random.randint(varmin,varmax)
a=random.randint(varmin,varmax)
a1=random.randint(varmin,varmax)
a2=random.randint(varmin,varmax)
b=random.randint(varmin,varmax)
b1=random.randint(varmin,varmax)
b2=random.randint(varmin,varmax)


x=linspace(0,2*pi,100)

y=k*np.cos(x-a)+b
y1=k1*np.cos(x-a1)+b1
y2=k2*np.cos(x-a2)+b2

plt.plot(x,y,'r-',x,y1,'b--', x, y2,'g')
plt.title('График y(x)')
plt.xlabel('Значения x')
plt.ylabel('Значения y')
plt.show()
```

3. Задание (в программе)
Напишите код, который будет переводить полярные координаты в декартовы.
Напишите код, который будет рисовать график окружности в полярных координатах.
Напишите код, который будет рисовать график отрезка прямой линии в полярных координатах.

```python
#код, который будет переводить полярные координаты в декартовы. 

def pol2cart(rho, phi):
    x = round(rho * math.cos(math.radians(phi)),3)
    y = round(rho * math.sin(math.radians(phi)),3)
    return(x,y)

rho=float(input("Введите значение радиуса rho: "))
phi=float(input("Введите значение phi в градусах: "))

print("Для полярных координат x = ", rho, "и y = ", phi, "декартовы координаты = ", pol2cart(rho, phi))
```
