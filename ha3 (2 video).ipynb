Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
import math
```

```python
x=float(input('введите координату Х вектора = '))
y=float(input('введите координату Y вектора = '))
z=float(input('введите координату Z вектора = '))
```

Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.

```python
import matplotlib.pyplot as plt
import decimal
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
xmin = -20
xmax = 20
dx = 0.2

x = np.around(np.arange(xmin, xmax, dx), decimals=4)

y = 1 / x

plt.plot(x ,y, color='green')
plt.title('График y(x) - гипербола')
plt.xlabel('Значения x')
plt.ylabel('Значения y')

plt.show()

#альтернатива - можно реализовать введение параметров масштаба x через ввод пользователя
```

```python
from math import pi

a=1    #radius 

t = np.linspace(0, 2*pi, 100)
x=a*np.cos(t)
y=a*np.sin(t)

plt.plot( x, y, color='green' )
plt.title('График y(x) -  окружность')
plt.xlabel('Значения x')
plt.ylabel('Значения y')
plt.axis('equal')
plt.show()

#альтернатива - можно реализовать введение  радиуса через ввод пользователя
```

```python
from math import pi
u=1    #x-position of the center
v=0.5    #y-position of the center
a=2    #radius on the x-axis
b=1.5    #radius on the y-axis

t = np.linspace(0, 2*pi, 100)
x=u+a*np.cos(t)
y=v+b*np.sin(t)

plt.plot( x, y, color='green' )
plt.title('График y(x) - эллипс')
plt.xlabel('Значения x')
plt.ylabel('Значения y')
plt.show()

#альтернатива - можно реализовать введение параметров через ввод пользователя
```
